```python
import os 
import numpy
import pandas
import json
import yaml
import re
```

```python
filepath="../main/java/me/matl114/logitech"
absPath=os.path.abspath(".")
display(absPath)
filepath = os.path.abspath(filepath)
os.chdir( filepath)
ds= os.walk(".\\")
filelist=[]
for fpath,dir_list,fil_list in list(ds):
    for file in fil_list:
        filelist.append(os.path.join(fpath,file))
os.chdir(absPath)
display(filelist)
```

```python
def spawn_key(fpath):
    fpath=fpath[2:].removesuffix(".java")
    fpath=fpath.split("\\")
    return fpath
yml_keys=[spawn_key(f) for f in filelist]
display(yml_keys)
```

```python
def extract_strings_from_java_file(file_path):
    os.chdir(filepath)
    # 读取Java文件内容
    with open(file_path, "r") as file:
        java_code = file.read()
    java_code = re.sub(r"//.*", "", java_code)
    #print(java_code)
    os.chdir(absPath)
    string_splits=list(java_code.replace("\\\"","$$$").replace("'","\"").split("\""))
    string_constants = [
        string_splits[i].replace("$$$", '\\"') for i in range(1, len(string_splits), 2)
    ]
    # 判断并提取符合链接形式的字符串常量
    other_strings = []
    for string_constant in string_constants:
        if not re.match(
            r"^[a-zA-Z]+\.[a-zA-Z]+(\.[a-zA-Z]+)*$", string_constant
        ) and re.search(r"[\u4e00-\u9fa5]", string_constant):
            other_strings.append(string_constant)

    return other_strings
display(extract_strings_from_java_file(filelist[0]))
```

```python
with open("./output.yml","r",encoding="utf-8") as fp:
    yaml_dict = yaml.load(fp, Loader=yaml.SafeLoader)
    if yaml_dict is None:
        yaml_dict={ }
lan_type="zh_CN"
yaml_dict[lan_type]={}
for f in filelist:
    key_path=spawn_key(f)
    root=yaml_dict[lan_type]
    for key in key_path:
        if key not in root.keys():
            root[key]={}
        root=root[key]
    root["enable"]=1
    for index,string in enumerate(extract_strings_from_java_file(f)):
        root[str(index)]={"lan":string,"key": '.'.join(key_path)+f".{str(index)}"}
display(yaml_dict)
with open("./output.yml","w",encoding="utf-8") as fp:
    fp.write(yaml.dump(yaml_dict, allow_unicode=True))
```
